```python
#to see how many duplicates are there in the dataset
data.duplicated().sum()
```

```python
# to check the duplicates in the unique feature
data["Order ID"].duplicated().sum()
data["Row ID"].duplicated().sum()
data["Customer ID"].duplicated().sum()
data["Product ID"].duplicated().sum()
```

```python
# to drop the duplicated in the unique features
data=data.drop_duplicates(subset="Order ID")
data=data.drop_duplicates(subset="Row ID")
data=data.drop_duplicates(subset="Customer ID")
data=data.drop_duplicates(subset="Product ID")
```

```python
# to check the dropped duplicates in unique id
data["Order ID"].duplicated().sum()
data["Row ID"].duplicated().sum()
data["Customer ID"].duplicated().sum()
data["Product ID"].duplicated().sum()
```

**EDA- EXPLORATORY DATA ANALYSIS:**

Here i have done a segmentation for the data analysis now need to find the answers for the businees problem and going to visualizse the results in the plots

```python
display(data)
```

```python
# using groupby to get the aswers
#Which customer segment (Consumer, Corporate, Home Office) places the most orders?
data1=data.groupby("Segment")["Order ID"].count()
data1
```

```python
plt.figure(figsize=(7,5))
plt.bar(data1.index,data1.values,color="blue",label="Segment Sales")
plt.title("Segment wise Sales")
plt.xlabel("Segment")
plt.ylabel("Sales")
plt.grid(axis="y")
plt.legend()
plt.show()
```

**Observation:**

Here the plot clearly shows the segment wise highest sales and the highest sales achieved my the segment was **Consumer** there is huge difference compared to other segments


```python
#Which ship mode is used most frequently by customers?
data2=data.groupby("Ship Mode")["Order ID"].count()
data2
```

```python
plt.figure(figsize=(7,5))
plt.bar(data2.index,data2.values,color="cyan",label="Frequent ship mode")
plt.title("Ship Mode")
plt.xlabel("Mode")
plt.ylabel("Total Orders")
plt.grid(axis="y")
plt.legend()
plt.show()
```

**Observation:**

we can clearly see that this ship mode have been achieved highest orders is **Standard class** compared to other modes


```python
#Do different customer segments prefer different ship modes?
data3=data.groupby(["Segment","Ship Mode"])["Order ID"].count()
data3
```
